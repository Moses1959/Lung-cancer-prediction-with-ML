<a id="13"></a>
# <p style="padding:10px;background-color:#749F82;margin:0;color:black;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Prétraitement pour la classification</p>

```python
#import relevant libraries
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier #Importing Random Forest Classifier
```

```python
#features extraction
x = data.drop("LUNG_CANCER", axis = 1)
y = data['LUNG_CANCER']
#Standardization of datasets is a common requirement for many machine learning estimators implemented in scikit-learn;
#they might behave badly if the individual features do not more or less look like standard normally distributed data
scaler = StandardScaler()
x = scaler.fit_transform(x)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)  #splitting data with test size of 20%

print(f"Shape of training data : {x_train.shape}, {y_train.shape}")
print(f"Shape of testing data : {x_test.shape}, {y_test.shape}")
```

<a id="14"></a>
# <p style="padding:10px;background-color:#749F82;margin:0;color:black;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Logistic regression model</p>

```python
logreg = LogisticRegression()   #build our logistic model
logreg.fit(x_train, y_train)  #fitting training data
y_pred  = logreg.predict(x_test)    #testing model’s performance
print("Logistic regression Accuracy=",round(logreg.score(x_test, y_test),4)*100,"%")
```

<a id="15"></a>
# <p style="padding:10px;background-color:#749F82;margin:0;color:black;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Confusion Matrix</p>

```python
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
```

##### Avec la matrice de confusion on peut constater que le model a prédit avec succès 56cas sur 62

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

##### Avec le classification report on peut voir que :
#### L'accuracy est estimé a 90% qui veut dire qu'on a reussi a predire avec succés 56 cas sur 62 cas
#### Le recall pour les cas postifs est estimé a 98% et a 50% pour les cas négatifs, qui veut dire qu'on a réussi a predire 51 personnes cancreuses de toutes les personnes qui ont un cancer(52). 
#### La précision pour les cas postifs est estimé a 91% et a 83% pour les cas négatifs, qui veut dire qu'on a prédit 56 personnes cancéreuses et que 51 d'entre elles étaient vraiment malades

<a id="16"></a>

```python
n_neighbors=18
scores=[]
for k in range(1,18):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    y_pred=knn.predict(x_test)
    print('Accuracy for k=',k,'is:',round(accuracy_score(y_pred,y_test),2))
    scores.append(round(accuracy_score(y_pred,y_test),2))
```

```python
plt.plot(range(1,18),scores)
plt.xlabel('Value K for KNN')
plt.ylabel('Testing Accuracy')
```
